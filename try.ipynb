```python
# Window Creation, retrieve link from client
# https://realpython.com/python-gui-tkinter/
from bs4 import BeautifulSoup
import tkinter as tk

window = tk.Tk()
greeting = tk.Label(
    text="Hi! Wanna give me an EA post link?",
    fg="white", 
    bg="black",
    width = 100,
    height = 30
)
greeting.pack()

label = tk.Label(
    text = "Insert Link Below",
    fg = "white",
    bg = "black"
)

entry = tk.Entry(
    fg="black",
    bg="white",
    width = 100
)

label.pack()
entry.pack()
link = entry.get()
soup = BeautifulSoup(link)
soup.find()
window.mainloop()
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
nltk.download('punkt')
text =  "In Andean Lives, Gregorio and his wife Asunta detail their life \
    as indigenous people living in Peru, recalling the many struggles they \
    had to endure as a result of their ethnicity and/or financial burdens, \
    as well as the changes in their life that shape their outlook on the \
    morality of others. While Gregorio lists the many specific instances of \
    racism and dishonesty he suffered, from being thrown in prison for \
    beef he did not steal, to getting thrown out of his living spaces \
    without good reason or notice, Asunta shares her difficulties \
    as a woman, which includes her husbands mistreating her as \
    well as her helplessness as an Indigenous woman without much opportunity. \
    As the stories of Asunta and Gregorio routinely includes violent or \
    disturbing details, the book is perfect for young adults like \
    myself, as the purpose of the book is to remind the reader \
    that there are so many other stories that are going untold, of people \
    who work just as hard as Gregorio and Asunta. I believe that it is \
    remarkable that this story is being read and discussed in my class, \
    as every adult needs to read a book such as this one for them to \
    understand the perspectives of those not as fortunate as themselves. \
    As Quechuans typically do not read or write, the perspectives of \
    Gregorio and Asunta become so much more valuable, since the Andeans’ \
    illiteracy and poverty-stricken lives are one of the many ways the \
    indigenous people in Peru are not able to stand up for themselves when \
    they are cheated, lied to, and mistreated by others. Learning through \
    experience is said to be the best way of learning, and with the powerful \
    stories in this book, I really felt like I could imagine both Asunta and \
    Gregorio in the situations they shared, and after finishing the book I \
    can sympathize more with the Andeans. I hope that others who read the \
    book will feel the same, and will be less selfish, and more open-minded \
    as a result, as we are more fortunate living in America than we likely \
    will ever fully realize. The quotes from both Asunta and Gregorio had a \
    hold on me that can only come from reading personal stories such as this. \
    Learning of the traditions and occasions of the people in the Andes in \
    class helped me from an outsider’s perspective looking in, while the book \
    threw us in headfirst, as if we were living and watching alongside \
    Gregorio and Asunta. What I learned in class was more like studying a \
    broad overview of the Andeans, which was a perfect complement to the book. \
    With such a straightforward and honest way of telling stories, the quotes \
    really are a slap in the face, and the personal element was just what was \
    necessary to gain a fuller perspective. Throughout reading this book, I \
    was happy to be able to point out specific details that we had learned \
    about in class: the unfamiliarity with cars, the birth difficulties, and \
    the fears of a drunk husband. My understanding of the book was enhanced \
    by what we learned about in class, and vice versa. Noteworthy instances \
    of the book are plentiful, as several moments shared in the book are much \
    different from anything that I’ve experienced in the United States. If I \
    had to pick only a few, one of the most powerful instances of the book to \
    me was Gregorio’s story about the stolen beef. While staying at a friend’s \
    house, Gregorio watched a cow get stolen, cooked, and enjoyed by his \
    friends. Later, a judge came and assured Gregorio that as long as he told \
    the truth, he wouldn’t be put in prison. He insisted so many times, in \
    fact, that Gregorio finally admitted that he had had a sip of the broth \
    that the beef was cooked in."
stopWords = set(stopwords.words("english"))
words = word_tokenize(text)

freqTable = dict()

for word in words :
    word = word.lower()
    if word in stopWords :
        continue
    if word in freqTable :
        freqTable[word] += 1
    else :
        freqTable[word] = 1
print(freqTable)

#create dict for score of each sentence
sentences = sent_tokenize(text)
sentenceValue = dict()

for sentence in sentences:
    for word, freq in freqTable.items() :
        if word in sentence.lower() :
            if sentence in sentenceValue :
                sentenceValue[sentence] += freq
            else :
                sentenceValue[sentence] = freq

sumValues = 0
for sentence in sentenceValue :
    sumValues += sentenceValue[sentence]

average = int(sumValues / len(sentenceValue))

summary = ""

for sentence in sentences:
    if (sentence in sentenceValue) and (sentenceValue[sentence] > (1.2 * average)) :
        summary += sentence

print(summary)
```
